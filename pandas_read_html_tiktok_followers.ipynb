# Most followed TikTok (read html table from Wikipedia)
https://en.wikipedia.org/wiki/List_of_most-followed_TikTok_accounts

```python
import pandas as pd
import numpy as np
print(f'pandas version: {pd.__version__}')
```

```python
from datetime import datetime
str(datetime.now())
```

```python
dfs=pd.read_html('https://en.wikipedia.org/wiki/List_of_most-followed_TikTok_accounts',
                 match='Description', header=0)
len(dfs)
```

```python
df=dfs[0]
df
```

```python
df.columns
```

```python
df.columns=df.columns.str.replace(r'\[\w+\]', '', regex=True)
df.columns
```

```python
df['Brand Account']=df['Brand Account'].replace(np.nan, 'x')
```

```python
df
```

```python
df.info()
```

```python
df.rename(columns={'Followers(millions)': 'Followers_mil', 'Likes(millions)':'Likes_mil'}, inplace=True)
```

```python
df['Rank']=pd.to_numeric(df['Rank'], errors='coerce')
df['Followers_mil']=pd.to_numeric(df['Followers_mil'], errors='coerce')
df['Likes_mil']=pd.to_numeric(df['Likes_mil'], errors='coerce')
```

```python
df.info()
```

```python
df['as_of']=str(pd.Timestamp.now().date())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
plt.figure(figsize=(7, 12))
sns.barplot(x='Followers_mil', y='Owner', 
            data=df.sort_values(by='Followers_mil', ascending=False),
            color='.75')
```

```python
colors = ['deepskyblue' if s in ['The Rock', 'BTS'] else '.75' for s in df['Owner']]
colors
```

```python
plt.figure(figsize=(7, 12))
sns.barplot(x='Followers_mil', y='Owner', 
            data=df.sort_values(by='Followers_mil', ascending=False),
            palette=colors)
```
