## AND with lookahead ?=

````python
df[df['Actors'].str.contains('.*Downey.*Evans', case=False)]
````

````python
df[df['Actors'].str.contains('.*Evans.*Downey', case=False)]
````

````python
df[df['Actors'].str.contains('(.*Evans.*Downey)|(.*Downey.*Evans)', case=False)]
````

````python
df[df['Actors'].str.contains('(?=.*Downey)(?=.*Evans)', case=False)]
````

````python
df[df['Actors'].str.contains('(?=.*Evans)(?=.*Downey)', case=False)]
````

## quantifiers
```
- *, ?, +
- a* -> a{0,}
- a+ -> a{1,}
- a? -> a{0,1}
```

````python
df[df['Title'].str.contains(r'\bi{2}\b', case=False)]
````

````python
df[df['Title'].str.contains(r'i{0,}', case=False)]
````
