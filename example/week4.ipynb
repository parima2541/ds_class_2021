```python
import pandas as pd
```

```python
df=pd.read_csv('https://github.com/prasertcbs/ds_class_2021/raw/week4/2602454S20202.txt', 
               names=['id', 'name_th', 'name_en', 'gender'], 
               usecols=[0, 1, 2, 3])
df
```

```python
df.columns
```

```python
df['name_th'].str.strip().str.extract(r'(น\.ส\.|นาย)(\w+)\s+(\w+)')
```

```python
dt=df['name_th'].str.strip().str.extract(r'(?P<title>น\.ส\.|นาย)(?P<fname>\S+)\s+(?P<lname>\S+)') # \S not \s , \D not \d
dt
```

```python
df['name_en'] # beware of trailing spaces
```

```python
df['name_en'].str.extract(r'(.+)\s+(.+)\s+(.+)') # \w alphanumeric
```

```python
df['name_en'].str.strip().str.extract(r'(?P<title_en>.+)\s+(?P<fname_en>.+)\s+(?P<lname_en>.+)') # \w alphanumeric
```

```python
df['name_en'].str.strip().str.extract(r'(?P<title_en>Miss|Mr\.)\s+(?P<fname_en>.+)\s+(?P<lname_en>.+)') # \w alphanumeric
```

```python
df['name_en'].str.strip().str.extract(r'(\w+)\s+(\w+)\s+(\w+)') # \w alphanumeric
```

```python
df['name_en']=df['name_en'].str.strip()
df['name_en']
```

```python
p=df['name_en'].str.split(r'\s+', expand=False)
p # series
```

```python
de=df['name_en'].str.strip().str.split(r'\s+', expand=True)
de 
```

```python
de.columns=['title_en', 'fname_en', 'lname_en']
de
```

```python
df.columns
```

```python
pd.concat([df[['id', 'gender']], dt, de], axis=1)
```
