```python
df['en']=df['item'].str.findall(r'\w+(?= \()') # ตัวอักษรที่อยู่ก่อน (
df
```

## lookahead (?=) and lookbehind (?<=)

```python
df['item'].str.findall(r'(?<=\().*(?=\))') # ตัวอักษรที่อยู่ใน ()
```

```python
df['item'].str.findall(r'(?<== ).*') # ตัวอักษรที่อยู่หลัง =space
```

```python
df['item'].str.extract(r'(?P<en>.*)\s+\((?P<pronunciation>.*)\)\s+\=\s+(?P<th>.*)')
```
